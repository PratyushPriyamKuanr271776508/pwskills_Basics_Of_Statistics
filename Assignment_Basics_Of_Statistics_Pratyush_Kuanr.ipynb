### 1. Types of Data

#### Qualitative (Categorical) Data:

- **Definition**: Non-numerical data used to classify or categorize objects.
- **Examples**:
  - Nominal: Gender (Male, Female), Blood type (A, B, AB, O).
  - Ordinal: Education level (High school, Bachelor’s, Master’s, Doctorate).

#### Quantitative (Numerical) Data:

- **Definition**: Numerical data representing measurable quantities.
- **Examples**:
  - Interval: Temperature in Celsius or Fahrenheit (no true zero).
  - Ratio: Height, Weight, Age (has a true zero).

### 2. Measures of Central Tendency

#### Mean:

- **Definition**: The average of all values.
- **Formula**: Mean = (Sum of all values) / (Number of values)
- **Example**: The average score of students in a test.
- **Use Case**: When data is symmetric without outliers.

#### Median:

- **Definition**: The middle value in an ordered dataset.
- **Example**: Median house price in a neighborhood.
- **Use Case**: When data has outliers or is skewed.

#### Mode:

- **Definition**: The most frequently occurring value(s) in a dataset.
- **Example**: Most common shoe size sold in a store.
- **Use Case**: For categorical data or identifying peaks in data.

### 3. Concept of Dispersion

- **Definition**: Measures the spread or variability in a dataset.
- **Variance**: Average squared deviation from the mean.
  - Formula: Variance = (Sum of squared differences from the mean) / (Number of values)
- **Standard Deviation**: Square root of variance, gives spread in original units.
  - Formula: Standard Deviation = Square root of Variance
- **Example**: Exam scores of students where a higher standard deviation indicates more variability.

### 4. Box Plot

- **Definition**: A graphical summary of data distribution.
- **Features**:
  - Median (central line).
  - Interquartile Range (IQR): Box length (Q3 - Q1).
  - Whiskers: Extend to 1.5 * IQR.
  - Outliers: Points beyond whiskers.
- **Use**: Identifies skewness, spread, and outliers.

### 5. Role of Random Sampling

- **Definition**: Selecting a subset from a population where each member has an equal chance of selection.
- **Importance**:
  - Reduces bias.
  - Ensures representativeness.
  - Enables inference about the population.
- **Example**: Polling a random group of voters to predict election results.

### 6. Skewness

- **Definition**: A measure of data asymmetry.
- **Types**:
  - Positive Skew: Long tail on the right.
  - Negative Skew: Long tail on the left.
- **Effect**:
  - Mean shifts towards the tail.
  - Median remains more robust to skewness.

### 7. Interquartile Range (IQR)

- **Definition**: Difference between the 75th percentile (Q3) and 25th percentile (Q1).
- **Formula**: IQR = Q3 - Q1
- **Use in Outlier Detection**:
  - Lower Bound: Q1 - 1.5 * IQR
  - Upper Bound: Q3 + 1.5 * IQR
  - Example: Finding outliers in income distribution.

### 8. Binomial Distribution

- **Conditions**:
  - Fixed number of trials (n).
  - Each trial has two outcomes (success or failure).
  - Constant probability of success (p).
  - Independent trials.
- **Example**: Flipping a coin 10 times and counting heads.

### 9. Normal Distribution and Empirical Rule

- **Properties**:
  - Symmetric and bell-shaped.
  - Mean = Median = Mode.
  - Defined by mean and standard deviation.
- **Empirical Rule**:
  - 68% of data within 1 standard deviation from the mean.
  - 95% of data within 2 standard deviations from the mean.
  - 99.7% of data within 3 standard deviations from the mean.
- **Example**: Heights of people.

### 10. Poisson Process

- **Definition**: Models the number of events occurring in a fixed interval of time/space.
- **Example**: Number of calls received at a call center in an hour.
- **Formula**: P(X = k) = (e^(-lambda) * lambda^k) / k!
  - lambda: Average rate.
  - k: Number of events.
- **Calculation**: If lambda = 5 calls/hour, probability of exactly 3 calls is:
  P(X = 3) = (e^(-5) * 5^3) / 3!.

### 11. Random Variables

- **Definition**: Variables representing outcomes of a random process.
- **Types**:
  - Discrete: Countable outcomes (e.g., dice roll).
  - Continuous: Infinite outcomes within a range (e.g., height).

### 12. Covariance and Correlation

- **Example Dataset**:
  - X: [2, 4, 6, 8]
  - Y: [5, 10, 15, 20]

- **Covariance**:
  - Formula: Cov(X, Y) = (Sum of (x - mean of X) * (y - mean of Y)) / Number of values
  - Result: Positive covariance indicates a direct relationship.

- **Correlation**:
  - Formula: Correlation(X, Y) = Covariance(X, Y) / (Standard deviation of X * Standard deviation of Y)
  - Result: Correlation = 1 indicates a perfect linear relationship.
